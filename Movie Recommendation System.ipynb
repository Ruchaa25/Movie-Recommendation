## Predict Function

```python
def predict_movie(movie_name):
    #selecting the movie from moviemat
    movie_user_rating= moviemat[movie_name]
    
    #finding the correlation
    similar_to_movie = moviemat.corrwith(movie_user_rating)
    
    #make df and drop nan values
    corr_movie = pd.DataFrame(similar_to_movie, columns=['correlation'])
    corr_movie.dropna(inplace=True)
    
    #joining with reviews and sort
    corr_movie = corr_movie.join(ratings['num_of_reviews'])
    predictions= corr_movie[corr_movie['num_of_reviews']>100].sort_values(by= 'correlation', ascending=False)
    
    return predictions
```

```python
predictions = predict_movie("Titanic (1997)")
```
